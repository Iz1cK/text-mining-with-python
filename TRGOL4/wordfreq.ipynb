```python
# Word frequencies
from nltk.book import *

sent3
fdist1=FreqDist(sent3)
fdist1
#print(fdist1)
fdist1.tabulate()

text1[:20]
fdist2 = FreqDist(text1)
print(fdist2)
fdist2.most_common(20)
fdist2['whale']
```

```python
# Only one word, whale, is slightly informative! It occurs over 900 times. 
# What proportion of the text is taken up with non-informative words? We can generate a cumulative frequency plot for these words, 
# These 50 words account for nearly half the book!

fdist2.plot(500)
#fdist2.plot(50, cumulative=True)

# Excercise: try this for text2
```

```python
fdist2.most_common()

# words that occur only once:
# In corpus linguistics, a hapax legomenon is a word or an expression that occurs only once within a context
#fdist2.hapaxes()

# what about hapaxes due to case?
text1l = [w.lower() for w in text1]
fdist3 = FreqDist(text1l)
#fdist3.hapaxes()

# what are the differences?
diff = set(fdist2.hapaxes()) - set(fdist3.hapaxes())
diff
len(fdist2.hapaxes())
len(diff)
```
